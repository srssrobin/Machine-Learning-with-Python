# **4. Dummy Variable/Categorical Variable Regression**

We'll use One Hot Encoding technique to perform dummy variable regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder as le
from sklearn.preprocessing import OneHotEncoder as ohc
from sklearn.compose import ColumnTransformer

```

```python
url = 'https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/5_one_hot_encoding/Exercise/carprices.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Use the pd.get_dummies() trick; it's a lot easier.
# Create a dummy variable column
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
# Merge the dummy column with the main column
merged_df = pd.concat([df, dummies], axis='columns')
merged_df.head()
```

```python
# Drop the original column, and one of the dummy variable columns to avoid Dummy Variable Trap
# dropped Mercedez Benz C class 'cause it looks annoying as hell, as well as the original Car Model column
merged_df.drop(['Car Model', 'Mercedez Benz C class'], axis='columns', inplace=True)
merged_df.head()
```

```python
merged_df.head()
```

```python
# Create X and Y variables and fit linear regression
X = merged_df.drop('Sell Price($)', axis='columns')
y = merged_df['Sell Price($)']
model = LinearRegression()
model.fit(X, y)
```

```python
# Check regression intercept and coefficients
model.intercept_, model.coef_
```

```python
# Ex 1: Predict Benz with 4 years, mileage of 45000
predict_1 = (model.predict([[45000, 4, 0, 0]]))
predict_1[0]

```

```python
# Ex 2: Prdict with BMW and 86,000 mileage
predict_2 = (model.predict([[86000, 7, 0, 1]]))
predict_2[0]
```
