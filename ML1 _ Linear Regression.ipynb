# **1. Linear Regression**

## **1. Linear Regression Overview**
Linear Regression is a statistical method used to model the relationship between a dependent variable $y$ and one or more independent variables $X$. In the simplest case, a Simple Linear Regression involves just one predictor, with the model defined as $y = \beta_0 + \beta_1 x + \epsilon$, where $beta_0$ is the intercept, $beta_1$ is the slope coefficient, and $epsilon$ represents the error term (assumed to be normally distributed with mean 0). In the case of Multiple Linear Regression, where multiple predictors are present, the model extends to $y = X\beta + \epsilon$, where $X$ is a matrix of predictors, $beta$ is a vector of coefficients, and $epsilon$ is a vector of errors. The goal of linear regression is to find the optimal set of $beta$ coefficients that minimize the difference between the observed and predicted values of $y$.

### **Ordinary Least Squares (OLS) Estimation**
The most common method to estimate the coefficients in a linear regression model is the Ordinary Least Squares (OLS) method. OLS seeks to minimize the sum of squared residuals (i.e., the differences between the observed values and the values predicted by the model). Mathematically, this is achieved by minimizing the objective function $S(\beta) = (y - X\beta)^T(y - X\beta)$. The residual sum of squares $(RSS)$ is a quadratic function of $\beta$, and finding its minimum involves taking the derivative of $S(\beta)$ with respect to $beta$, setting it equal to zero, and solving for $beta$. This results in the OLS estimator $\hat{\beta} = (X^TX)^{-1}X^Ty$, provided that $X^TX$ is invertible.

### **Matrix Algebra Derivation of OLS**
To derive the OLS estimator using matrix algebra, start with the model $y = X\beta + \epsilon$. The vector of residuals is given by $e = y - X\beta$. The OLS objective is to minimize the residual sum of squares, which can be expressed in matrix form as $e^Te = (y - X\beta)^T(y - X\beta)$. Expanding this, we get $y^Ty - 2\beta^TX^Ty + \beta^TX^TX\beta$. To minimize this expression, take the partial derivative with respect to $beta$ and set it to zero: $-2X^Ty + 2X^TX\beta = 0$. Solving for $beta$ yields the normal equation $X^TX\beta = X^Ty$. Assuming $X^TX$ is invertible, the solution is $\hat{\beta} = (X^TX)^{-1}X^Ty$. This is the OLS estimator, which provides the best linear unbiased estimate of $beta$ under the Gauss-Markov assumptions (e.g., $linearity$, $independence$, $homoscedasticity$, and no $multicollinearity$).

## **2. Derivation of OLS Using Matrix Algebra:**

In this notebook, we will derive the Ordinary Least Squares (OLS) estimator step by step using matrix algebra. Our goal is to find the coefficient vector $$\hat{\beta}$$ that minimizes the sum of squared residuals in a linear regression model.


### **1. Linear Regression Model**

The multiple linear regression model can be expressed in matrix form as:

$$
y = X\beta + \epsilon
$$

where:
* $y$ is an $n \times 1$ vector of the dependent variable.
* $X$ is an $n \times p$ matrix of independent variables (including a column of ones for the intercept).
* $\beta$ is a $p \times 1$ vector of coefficients.
* $\epsilon$ is an $n \times 1$ vector of error terms.


### **2. Define the Residuals**

The residuals $e$ are defined as the difference between the observed values $y$ and the predicted values $\hat{y}$:

$$
e = y - X\hat{\beta}
$$

Our goal is to minimize the sum of squared residuals (RSS):

$$
RSS = e^T e = (y - X\hat{\beta})^T (y - X\hat{\beta})
$$


### **3. Expand the Quadratic Expression**

Let's expand the quadratic form of the RSS:

$$
RSS = (y - X\beta)^T (y - X\beta)
$$

Expanding this, we get:

$$
RSS = y^T y - y^T X\beta - \beta^T X^T y + \beta^T X^T X \beta
$$

Since $y^T X\beta$ and $\beta^T X^T y$ are scalars, they are equal, so we can simplify:

$$
RSS = y^T y - 2\beta^T X^T y + \beta^T X^T X \beta
$$


### **4. Take the Derivative with Respect to $\beta$**

To minimize the RSS, we take the derivative of the RSS with respect to $\beta$ and set it to zero:

$$
\frac{\partial RSS}{\partial \beta} = -2X^T y + 2X^T X \beta = 0
$$

This simplifies to:

$$
X^T X \beta = X^T y
$$


### **5. Solve for $\hat{\beta}$**

We can solve for $\beta$ by multiplying both sides by the inverse of $X^T X$ (assuming it is invertible):

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

This is the OLS estimator for the coefficient vector $\beta$.

### **6. Summary of the OLS Estimator**

The Ordinary Least Squares (OLS) estimator is given by:

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

This formula provides the best linear unbiased estimate (BLUE) of the coefficients, minimizing the sum of squared residuals between the observed and predicted values.

### Key Assumptions:
1. **Linearity**: The relationship between the dependent and independent variables is linear.
2. **Independence**: The observations are independent of each other.
3. **Homoscedasticity**: The error terms have constant variance.
4. **No multicollinearity**: The independent variables are not perfectly collinear.
5. **Normality of errors** (optional for small sample sizes): The error terms are normally distributed.


## **3. Gauss Markov Assumptions:**

In this notebook, we outline the Gauss-Markov assumptions using matrix notation, which are the basis for the Ordinary Least Squares (OLS) estimator to be the **Best Linear Unbiased Estimator (BLUE)**.

Given the linear regression model:

$$
y = X\beta + \epsilon
$$

where:
- $ y $ is an $ n \times 1 $ vector of the dependent variable.
- $ X $ is an $ n \times p $ matrix of independent variables (including a column of ones for the intercept).
- $ \beta $ is a $ p \times 1 $ vector of coefficients.
- $ \epsilon $ is an $ n \times 1 $ vector of error terms.

Here are the Gauss-Markov assumptions expressed in matrix form:


#### **1. Linearity of the Model**

The relationship between \( y \) and \( X \) is linear:

$$
y = X\beta + \epsilon
$$

This assumption means the model is correctly specified as a linear combination of the independent variables.


#### **2. Full Rank Condition (No Perfect Multicollinearity)**

The matrix $ X $ must have full column rank, which means:

$$
\text{rank}(X) = p
$$

This condition ensures that $ (X^T X)^{-1} $ exists, and without it, the coefficients cannot be uniquely estimated.

#### **3. Exogeneity (Zero Mean of Errors)**

The error terms have an expected value of zero, given the independent variables:

$$
\mathbb{E}(\epsilon | X) = 0
$$

This implies that the conditional mean of $ y $, given $ X $, is:

$$
\mathbb{E}(y | X) = X\beta
$$

#### **4. Homoscedasticity (Constant Variance of Errors)**

The error terms have constant variance and are uncorrelated with each other:

$$
\text{Var}(\epsilon | X) = \sigma^2 I_n
$$

where:
- $ \sigma^2 $ is a constant.
- $ I_n $ is the $ n \times n $ identity matrix.

This means the covariance matrix of the errors is:

$$
\text{Cov}(\epsilon) = \sigma^2 I_n
$$

#### **5. No Autocorrelation (Independence of Errors)**

The error terms are uncorrelated with one another:

$$
\text{Cov}(\epsilon_i, \epsilon_j | X) = 0 \quad \text{for} \; i \neq j
$$

In matrix form, this is represented as:

$$
\text{Cov}(\epsilon) = \sigma^2 I_n
$$

which indicates that the errors are both homoscedastic and independent.

#### **6. Normality of Errors (Optional for BLUE, Required for Inference)**

For the purpose of statistical inference (hypothesis testing and confidence intervals), we assume that:

$$
\epsilon \sim \mathcal{N}(0, \sigma^2 I_n)
$$

This assumption is not required for the OLS estimator to be BLUE, but it is necessary for deriving the exact sampling distribution of the OLS estimator and conducting hypothesis tests.

#### Summary

If the first five assumptions are satisfied, the OLS estimator:

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

is the **Best Linear Unbiased Estimator (BLUE)**, meaning it has the lowest variance among all linear unbiased estimators.

By expressing the assumptions in matrix form, we can better understand how they ensure the reliability and efficiency of the OLS estimates.


## **4. Python Implementation:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

Load Dataset

```python
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv'

df  = pd.read_csv(url)

df.head(10)
```

Investigate dataset

```python
df.year.unique()
```

Visualize the data

```python
sns.regplot(x = 'year', y = 'per capita income (US$)', data = df)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Per Capita Income vs Year")
plt.show()
```

Fit Linear Regression model

```python
# Set up regression object
reg = LinearRegression()

# Fit regression model
reg.fit(df[['year']], df['per capita income (US$)'])

# Check regression coefficient and intercept
reg.coef_, reg.intercept_
```

Predict for Year 2025

```python
reg.predict([[2025]])
```
